## Импорты

```python
import random
from itertools import islice
from pprint import pprint

import wget
from corus import load_lenta
from tqdm import tqdm

from preprocessing import delete_stop_words, lemmatize, normalize, tokenize_by_sentence, tokenize_by_words
```

```python
random.seed(42)
```

## Скачиваем датасет

```python
dataset_url = 'https://github.com/yutkin/Lenta.Ru-News-Dataset/releases/download/v1.0/lenta-ru-news.csv.gz'
wget.download(dataset_url)
```

```python
path = 'lenta-ru-news.csv.gz'
records = load_lenta(path)
```

Оставим только текст из записей датасета

```python
news = (record.text for record in records)
```

## Препроцессинг данных

```python
def preprocess(text:str) -> str:
    sentences:str = tokenize_by_sentence(text)
    preprocessed_sentences:list[str] = []
    for sentence in sentences:
        normalized_text = normalize(sentence)
        tokens = tokenize_by_words(normalized_text)
        clean_tokens = delete_stop_words(tokens)
        lemmatized_tokens = lemmatize(clean_tokens)
        preprocessed_sentences.append(" ".join([token for token in lemmatized_tokens]))
    return preprocessed_sentences
```

Выведем 5 записей новостей до препроцессинга и после

```python
first_5_news = list(islice(news, 5))
```

```python
for news in first_5_news:
    print(f"ДО: \n {news}")
    print(f"ПОСЛЕ: \n {preprocess(news)}")
```
